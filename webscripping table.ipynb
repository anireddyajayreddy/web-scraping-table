```python
rows =table[0].find_all('tr')
```

```python
for i in rows:
    print(i)
```

```python
all_rows=[]
for i in rows:
    td = i.find_all('td')
    row=[i.text for i in td]
    all_rows.append(row)
```

```python
all_rows.pop(0)
```

```python
all_rows
```

```python
for i in all_rows:
  i[0] = i[0][:i[0].find("DistrictCases")]
```

```python
for i in all_rows:
  i[1] = i[1][:i[1].find(" ")]
  i[2] = i[2][:i[2].find(" ")]
  i[3] = i[3][:i[3].find(" ")]
  i[4] = i[4][:i[4].find(" ")] 
```
